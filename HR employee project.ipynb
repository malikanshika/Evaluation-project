```python
df['Attrition']=df['Attrition'].replace({'No':0,'Yes':1})
```

```python
df['OverTime']=df['OverTime'].map({'No':0,'Yes':1})
df['Gender']=df['Gender'].map({'Male':0,'Female':1})
```

```python
df['Over18']=df['Over18'].map({'Y':1,'No':0})
```

```python
from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders={}
for column in encoding_cols:
    label_encoders[column]=LabelEncoder()
    df[column]=label_encoders[column].fit_transform(df[column])
```

```python
df.info()
```

```python
X=df.drop(['Attrition'],axis=1)
y=df['Attrition'].values
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus=RandomOverSampler(random_state=42)
X_over,y_over=rus.fit_resample(X,y)
print(Counter(y_over))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_over,y_over,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve,roc_auc_score
```

```python
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
prediction=logreg.predict(X_test)
cnf_matrix=confusion_matrix(y_test,prediction)
print("Accuracy Score-",accuracy_score(y_test,prediction))
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier,ExtraTreesClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
svc = SVC()
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
sc1 = metrics.accuracy_score(y_pred,y_test)
print(sc1)
```

```python
gbmMod = GradientBoostingClassifier(max_depth = 3, n_estimators = 100, learning_rate = 0.05, random_state = 1)
gbmMod.fit(X_train,y_train)
y_pred = gbmMod.predict(X_test)
sc2 = metrics.accuracy_score(y_pred,y_test)
print(sc2)
```

```python
# Random Forest
RandFor = RandomForestClassifier(max_depth = 3, min_samples_split=2, n_estimators = 100, random_state = 1)
RandFor.fit(X_train, y_train)
y_pred = RandFor.predict(X_test)
sc3 = metrics.accuracy_score(y_pred,y_test)
print(sc3)
```

```python
ET=ExtraTreesClassifier()
ET.fit(X_train,y_train)
training_data_prediction=ET.predict(X_test)
sc4=metrics.accuracy_score(y_test,training_data_prediction)

print(sc4)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=4, random_state=0)
clf.fit(X_train, y_train)
predict = clf.predict(X_test)
sc5=accuracy_score(y_test,predict)
sc5
```
